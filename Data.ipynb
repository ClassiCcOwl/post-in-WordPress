```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

option = webdriver.ChromeOptions()
option.add_argument('headless')


driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()),options=option)
```

```python
def getMap(driver) -> list:
    ul_xpath = '//*[@id="position"]/div/ul'
    ul_element = driver.find_element(By.XPATH , ul_xpath)
    liList = ul_element.find_elements(By.TAG_NAME, 'li')
    liList = [li.text for li in liList]
    return "\n".join(liList)
```

```python
def fileGenerator(driver):
    b = ["\\", "/", "<", ">", "|", "*", "?", '"', ":"]
    
    product = driver.find_element(By.CLASS_NAME, "product")
    product_text = product.text
    
    summary = driver.find_element(By.CLASS_NAME, "product-summary")
    h2 = summary.find_element(By.TAG_NAME, "h2")
    title = h2.text
    for _ in b:
        title = title.replace(_ , " ")
    
    with open(f'{title}.txt', 'a', encoding="utf-8") as file:
        file.write(getMap(driver))
        file.write(product_text)
    
```

```python
import requests
from bs4 import BeautifulSoup as bs
cUrl = 'https://www.rpsi.ir/fa/genericdrug/item/7942'
b = ["\\", "/", "<", ">", "|", "*", "?", '"', ":"]

r= requests.get(cUrl)
soup = bs(r.content, 'html5lib')
summary = soup.find('div',class_='product-summary')
h2 = summary.find('h2').text

short = soup.find('div',class_="short-description")

product = soup.find_all('div', class_="content-box")[:-1]

for _ in b:
    h2 = h2.replace(_, " ")

with open (f'{h2}.txt', 'a', encoding="utf-8") as file:
    file.write(h2)
    for sh in short.contents:
        file.write(str(sh))
    for div in product:
        file.write(str(div))
```
