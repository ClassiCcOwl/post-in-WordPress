```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
```

```python
driver.get("https://hosseini1.ir/login?lang=en")
driver.maximize_window()
email_xpath = '//*[@id="login-input-email"]'
password_xpath = '//*[@id="login-input-password"]'
edit_btn_xpath = '//*[@id="app0"]/div/div[2]/div[3]/a'
red_btn_xpath = '//*[@id="carousel_pages"]/div[1]/div/div[2]/li/div/div[2]'
my_email = ""
my_password = ""

WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, email_xpath))).send_keys(my_email)
WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, password_xpath))).send_keys(my_password,Keys.ENTER)
WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, edit_btn_xpath))).click()
WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, red_btn_xpath))).click()
```

```python
def poster(driver, my_cat, title, text):
    cat_xpath = '//*[@id="s2id_filter-taxons"]/a'
    cat_xpath2 = '//input[starts-with(@class,"select2-input")]'
    add_btn_xpath = '//*[@id="tab-video"]/div/div/button'
    title_xpath = '//*[@id="title"]'
    code_xpath = '//i[@class="mce-ico mce-i-code"]'
    text_box_xpath = '//textarea[contains(@id,"mceu")]'
    submit_btn_xpath = '//div[contains(@class,"last")]//following::button[31]'

    post_btn_xpath = '//div[@class="mce-widget mce-btn mce-primary mce-abs-layout-item mce-first mce-btn-has-text"]'
    
    confirm_btn_xpath = '//button[starts-with(@class,"confirm")]'
    WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, cat_xpath))).click()
    print("cat seleceted")
    inp = driver.find_elements(By.XPATH, cat_xpath2)
    inp[-1].send_keys(my_cat,Keys.ENTER)
    print("cat typed")
    WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, add_btn_xpath))).click()
    print("add")
    from time import sleep
    sleep(.7)
    WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, title_xpath))).send_keys(title)
    print("title typed")
    WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, code_xpath))).click()
    print("sourse code")
    
    text_box = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, text_box_xpath)))
    text_box._parent.execute_script("""
    var elm = arguments[0], text = arguments[1];
    if (!('value' in elm))
      throw new Error('Expected an <input> or <textarea>');
    elm.focus();
    elm.value = text;
    elm.dispatchEvent(new Event('change'));
    """, text_box, text)

    WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, post_btn_xpath))).click()  
    
    print("text typed")
    
    WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, submit_btn_xpath))).click()
    print("su")
    WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, confirm_btn_xpath))).click()
    print("done")
```
